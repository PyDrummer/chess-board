```python
# Creating a chessboard grid 8x8x3
# Dont forget to run each line of code.
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Board:
    def __init__(self):
        self.grid = np.zeros((8, 8, 3))
        self.grid[1::2, ::2] = 1
        self.grid[::2, 1::2] = 1
        self.red = []
        self.blue = []

    def add_red(self, row, col):
        red = (1.0, 0.0, 0.0)
        self.grid[row][col] = red
        self.red = [row, col]

    def add_blue(self, row, col):
        blue = (0.0, 0.0, 1.0)
        self.grid[row][col] = blue
        self.blue = [row, col]

    def render(self):
        # grid[1::2, ::2] picks every other element starting from the first element on the row and for every second row of the matrix.
        # https://stackoverflow.com/questions/2169478/how-to-make-a-checkerboard-in-numpy
        plt.imshow(self.grid)

    def is_under_attack(self):
        # Diagnoal must have an even number between the spaces of the two queens
        # Returns a boolean if red is under attack by a blue piece horizontally, vertically or diagonally
        red_z = self.red[0] + 2
        blue_z = self.blue[0] + 2
        red_v = self.red[1] + 2
        blue_v = self.blue[1] + 2

        dy = abs(red_z - blue_z)
        dx = abs(red_v - blue_v)
        if red_z % blue_z == 0 or blue_z % red_z == 0:
            # print("True red_z %\ blue_z == 0 or blue_z %\ red_z == 0")
            return True
        elif (red_v % blue_v == 0 or blue_v % red_v == 0):
            # print("True at red_v %\ blue_v == 0 or blue_v %\ red_v == 0")
            return True
        else:
            return dx == dy
       
```

```python
# Diagonal True
b = Board()
b.add_blue(1, 1)
b.add_red(5, 5)
print(b.is_under_attack())
b.render()
```

```python
# Same row true
b = Board()
b.add_blue(2, 2)
b.add_red(2, 6)
print(b.is_under_attack())
b.render()
```

```python
# Same Column True
b = Board()
b.add_blue(3, 4)
b.add_red(5, 4)
print(b.is_under_attack())
b.render()
```

```python
# Show False
b = Board()
b.add_blue(3, 4)
b.add_red(1, 3)
print(b.is_under_attack())
b.render()
```
